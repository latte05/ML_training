```
#Install OpenCV 
import cv2
```

```
#import picture files
import urllib.request as req
url = "http://uta.pw/shodou/img/28/214.png"
req.urlretrieve(url, "test.png")

img = cv2.imread("test.png")
```

```
%matplotlib inline

import matplotlib.pyplot as plt

#OpenCV = BGR (Blue Green Red)
#matplotlib = RGB (Red Green Bue)
#it needs to convert from BGR to RGB to show correctly
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```
#write image
cv2.imwrite("out.png", img)
```

```
#change size
im2 = cv2.resize(img, (600, 300)) # (width, height)
plt.imshow(cv2.cvtColor(im2, cv2.COLOR_BGR2RGB))
plt.show()
```

```
#crop
im3 = img[150:450, 150:450]
im3 = cv2.resize(im3, (400, 400))
plt.imshow(cv2.cvtColor(im3, cv2.COLOR_BGR2RGB))
plt.show()
```
